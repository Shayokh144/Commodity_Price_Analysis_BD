```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getDataFrameFromCSV(filePath):
    df = pd.read_csv(filePath + '.csv')
    return df
```

```python
def createPlot(dataFrame, columnName, plotTitle = ""):
    colValues = dataFrame[columnName]
    dateValues = dataFrame['date']
    ypoints = np.array(colValues)
    xpoints = np.array(dateValues)
    f = plt.figure()
    f.set_figwidth(12)
    f.set_figheight(6)
    plt.plot(xpoints, ypoints, color='green')
    plt.title(plotTitle)
    plt.xlabel("days")
    plt.ylabel("price")
    plt.show()
```

```python
df = getDataFrameFromCSV('../../data/price_data/rice/rice')
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
print("Sample Data:")
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
print(df.head(2))
```

```python
itemName = "Nazirshail Rice Standard (Half Boiled)"
filteredDf = df.loc[(df.product_name == itemName) & (df.weight_value == 5)]
createPlot(filteredDf, 'price', itemName)
```
