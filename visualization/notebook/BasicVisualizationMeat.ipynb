```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getDataFrameFromCSV(filePath):
    df = pd.read_csv(filePath + '.csv')
    return df
```

```python
def createPlot(dataFrame, columnName, plotTitle = ""):
    colValues = dataFrame[columnName]
    dateValues = dataFrame['date']
    ypoints = np.array(colValues)
    xpoints = np.array(dateValues)
    f = plt.figure()
    f.set_figwidth(12)
    f.set_figheight(6)
    plt.plot(xpoints, ypoints, color='green')
    plt.title(plotTitle)
    plt.xlabel("days")
    plt.ylabel("price")
    plt.show()
```

```python
df = getDataFrameFromCSV('../../data/price_data/meat/meat')
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
print("Sample Data:")
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
print(df.head(10))
```

```python
def showLineGraph(dataFrame, item, weight):
    df = dataFrame
    itemName = item
    filteredDf = df.loc[(df.product_name == itemName) & (df.weight_value == weight)]
    createPlot(filteredDf, 'price', itemName)
```

```python
showLineGraph(df, "Beef Bone In ± 50 gm", 1)
```

```python
showLineGraph(df, "Mutton Bone in (Khashi) ± 50 gm", 1)
```

```python
def createMultilinePlot(dataFrames, columnNames, plotTitle = "", weight = "1", lineNames = []):
    f = plt.figure()
    f.set_figwidth(8)
    f.set_figheight(4)
    i = 0
    colors = ['green', 'blue']
    for dataFrame, columnName in zip(dataFrames, columnNames):
        #dataFrame.head(3)
        colValues = dataFrame[columnName]
        dateValues = dataFrame['date']
        ypoints = np.array(colValues)
        xpoints = np.array(dateValues)
        plt.plot(xpoints, ypoints, color=colors[i], label=lineNames[i])
        i += 1
        i %= 2
    #plt.title(plotTitle+ "\n unit:"+ weight)
    plt.xlabel("days")
    plt.ylabel("price BDT")
    plt.legend(loc='best')
    plt.title("unit 1 ltr")
    plt.show()
```

```python
def showMultipleLineGraph(dataFrame, items, weights, weightRaws):
    filteredDfs = []
    itemNames = []
    df = dataFrame
    for itemName, weight in zip(items,weights):
        itemNames.append(itemName)
        filteredDf = df.loc[(df.product_name == itemName) & (df.weight_value == weight)]
        filteredDfs.append(filteredDf)
    createMultilinePlot(filteredDfs, ['price','price'], itemNames, weightRaws, items)
```

```python
items = ["Beef Bone In ± 50 gm", "Mutton Bone in (Khashi) ± 50 gm"]
showMultipleLineGraph(df, items, [1,1], [ "1 kg","1 kg"])
```
